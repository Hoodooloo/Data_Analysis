### Task 3-Find out the per unit price from the data

```python
#a. Convert date time field into pandas time object
sale_df = pd.DataFrame(sales)
sale_df["Order Date"] = pd.to_datetime(sale_df["Order Date"])
sale_df["Ship Date"] = pd.to_datetime(sale_df["Ship Date"])
sale_df.info()
```

```python
# b. Find out the price per unit from the data and create a column for it in the same data
sales_df["Price"] = (sales_df["Sales"]/sales_df["Quantity"])
sales
```

### Task 4-Find out the monthly revenue and analyze the findings

```python
# a.Create a column with Year and month field only give a suitable name to it
sales_df["Month-Year"] = pd.DatetimeIndex(sales_df["Order Date"]).to_period('M')
sales_df
```

```python
# b.Create a separate dataset from the data which will have two columns ,
# one which is created in step a of this task and second the monthly sales /
# revenue
mr = pd.DataFrame(sales_df)  #monthly_revenue
mr_df = mr.groupby('Month-Year',as_index=False)["Sales"].sum()
mr_df
```

```python
# c.Plot this dataset
mr_df.plot(x='Month-Year', y='Sales')
```

### Task 5-Find out the yearly revenue and analyze the findings

```python
# a.Create a column with Year and month field only give a suitable name to it
sales_df["Year"] = pd.DatetimeIndex(sales_df["Order Date"]).to_period('Y')
sales_df.head()
```

```python
# b.Create a separate dataset from the data which will have two columns ,
# one which is created in step a of this task and second the yearly sales /
# revenue
    
yr = pd.DataFrame(sales_df)         # Yearly_revenue
yr_df = yr.groupby('Year',as_index=False)["Sales"].sum()
yr_df
```

```python
# c.Plot this dataset
yr_df.plot(kind='bar',x='Year', y='Sales')
```

### Task 6-Finding out the monthly growth rate and analyse the findings

```python
# a.Create a column in the monthly revenue dataset for monthly growth rate
# mgr = monthly growth rate
mgr = (((mr_df["Sales"].shift())-mr_df["Sales"])/mr_df["Sales"])*100
mr_df["Monthly Growth rate"] = mgr
mr_df['Monthly Growth rate'] = mr_df['Monthly Growth rate'].fillna(0)
mr_df
```

```python
# b.Find out the maximum monthly growth rate
mr_df.max()
```

```python
# c.Plot this findings
mr_df.plot(kind='bar',x='Month-Year', y='Monthly Growth rate')
```

### Task 11-Find out the most and least sold product category from the store

```python
# a.Value based
product = sales[['Category', 'Sales']]
product_df = product.groupby('Category',as_index=False)["Sales"].sum()
print(product_df)
print("\n")
max_product = product_df.max()
min_product = product_df.min()
print(str(max_product) + " \n " + str(min_product))
```

```python
# b.Quantity based
product = sales[['Category', 'Quantity']]
product_df = product.groupby('Category',as_index=False)["Quantity"].sum()
print(product_df)
print("\n")
max_product = product_df.max()
min_product = product_df.min()
print(str(max_product) + " \n " + str(min_product))
```
